# String Matching Algorithm

```python
class StringMatching():
    def __init__(self):
        self.checker = False

    def yes_or_no(self, S, Q):
        for i in range(len(Q)):
            print(f"{i+1}번째 매칭 대상 문자열에 대한 결과입니다.")
            for j in range(len(S)):
              p = re.compile(S[j]) 
              s = p.search(Q[i])
              if s:
                  print("YES")
              else:
                  print("NO")
            print('\n')

    def find(self, S, Q):
        self.checker = False

        for i in range(len(Q)):
            print(f"{i+1}번째 매칭 대상 문자열에 대한 결과입니다.")
            for j in range(len(S)):
              p = re.compile(S[j]) 
              s = p.search(Q[i])
              if s:
                  print(f"#pos = {s.start()}, pattern = {s.group()}")
              else:
                  pass
            print('\n')

        self.checker = True

    def findall(self, S, Q):
        for pat in S:  
            for txt in Q:
                M = len(pat) 
                N = len(txt) 
                
                for i in range(N-M+1):      

                    # 끝 문자 다르면 바로 한칸 이동
                    if pat[M-1] != txt[M-1+i]:
                        pass

                    # 끝 문자 같으면 비교 시작
                    else:
                        pat_idx = M-2 #i=0, pat_idx=1 / pat_idx=0
                        txt_idx = M-2+i #i=0, txt_idx=1 / txt_idx=0
                        while pat_idx >= 0: #pat 인덱스 처음까지 작동 
                            if pat[pat_idx] != txt[txt_idx]: #txt는 계속 이동한 값
                                break
                                # pat[1] == txt_idx[1] / pat[0] == txt[0]
                            pat_idx -= 1 # 패턴 비교인덱스 1씩 감소
                            txt_idx -= 1 # 텍스트 비교인덱스 1씩 감소
                            
                        if pat_idx == -1: # 비교 텍스트가 모두 같은 상황에 도착
                            print(f"pos = {txt_idx + 1}, pattern = {pat}") # 포지션 출력

    # find와 findall을 한번에 진행하는 메서드도 만들어 봤습니다.
    def find_and_findall(self, S, Q):
        find = self.find(S, Q)
        if self.checker:
            findall = self.findall(S, Q)
        return find, findall



# Driver Code
if __name__ == '__main__':
    # 클래스 객체 생성
    StringMatching = StringMatching()

    """
    [임의의 입력값]
    주석 값을 복사 붙여넣기 하거나, 주석 값 형식을 맞춰 임의의 값을 넣어도 됩니다.
    """
    # 문자열 집합 S의 원소 개수 : 9
    print("문자열 집합 S의 원소 개수 : ")
    S_num = input()

    # 문자열 집합 S : 'aaa', 'aaaabb', 'aabbcc', 'abb', 'bcc', 'bbcc', 'aabbccdd', 'aaabb', 'cccd'
    print("문자열 집합 S : ")
    S = input()
    
    # 매칭 대상 문자열 수 : 2
    print("매칭 대상 문자열 개수 : ")
    Q_num = input()

    # 매칭 대상 문자열들 : 'aaaabbaabbccdd', 'aaabaabbcccd'
    print(f"매칭 대상 문자열 : ")
    Q = input()

    # 입력값들을 원소별로 split 후 리스트화
    S = S.replace("'", '').split(", ")
    Q = Q.replace("'", '').split(", ")

    # 출력
    print('\n')
    print("=====================YES or NO 출력=====================\n")
    StringMatching.yes_or_no(S, Q) 
    print('\n')
    print("========첫번째로 매칭된 패턴의 시작포지션과 문자열 패턴 출력========\n")
    StringMatching.find(S, Q)
    print('\n')
    print("======가능한 모든 문자열 패턴의 시작 포지션과 문자열 패턴 출력======\n")
    StringMatching.findall(S, Q)
    print('\n')
    print("======find and find_all======\n")
    StringMatching.find_and_findall(S, Q)
```

```python
# Searching algorithm 
def search(pat, txt): 
    M = len(pat) 
    N = len(txt) 
  
    # A loop to slide pat[] one by one */ 
    for i in range(N - M + 1): 
        j = 0
          
        # For current index i, check  
        # for pattern match */ 
        while(j < M): 
            if (txt[i + j] != pat[j]): 
                break
            j += 1
  
        if (j == M):  
            print("Pattern found at index ", i) 
  
# Driver Code 
if __name__ == '__main__': 
    txt = "AABAACAADAABAAABAA"
    pat = "AABA"
    search(pat, txt) 
```
